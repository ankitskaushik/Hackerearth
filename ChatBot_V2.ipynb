```python
def info(text):
    #Removing punctuation
    try:
        text=re.sub(r"[^\w\s]|\d+","", text)
    except:
        pass
    #Converting to lowercase text
    text = text.lower()
    
    #Tokenize response
    try:
        text = word_tokenize(text)
    except:
        pass
    
    #Stop Word Removal
    stop_words = stopwords.words('english')
    stop_words = set(stop_words)
    for i in range(len(text)):
        filtered=[]
        try:
            for j in text:
                if j not in stop_words:
                    filtered.append(j)
                    
                else:
                    pass
            text = filtered
        except:
            continue
    
    #Lexicon Normalization
    lem = WordNetLemmatizer()
    for i in range(len(text)):
        x,y='',[]
        try:
            for j in text:
                x=lem.lemmatize(j, pos='v')
                y.append(x)
            text = y
        except:
            pass
        
    return text
```
